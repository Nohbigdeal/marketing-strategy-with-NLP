```python
#í•¨ìˆ˜ ì„ ì–¸
def add_dataframe(name,category,reviews,stars,cnt):  #ë°ì´í„° í”„ë ˆì„ì— ì €ì¥
    #ë°ì´í„° í”„ë ˆì„ìƒì„±
    df1=pd.DataFrame(columns=['type','category','review','star'])
    n=1
    if (cnt>0):
        for i in range(0,cnt-1):
            df1.loc[n]=[name,category,reviews[i],stars[i]] #í•´ë‹¹ í–‰ì— ì €ì¥
            i+=1
            n+=1
    else:
        df1.loc[n]=[name,category,'null','null']
        n+=1    
    return df1

def save():
    if not os.path.exists('ë„¤ì´ë²„ ì‡¼í•‘/output1.csv'):
        df1.to_csv('ë„¤ì´ë²„ ì‡¼í•‘/output1.csv', encoding='utf-8-sig', mode='w')
    else:
        df1.to_csv('ë„¤ì´ë²„ ì‡¼í•‘/output1.csv',encoding='utf-8-sig', mode='a',header=False)
```
